```python
import scipy.stats  as sas
pos=np.split(np.arange(1963,1998,3)[::-1],2)
pos_span=np.arange(1963,1998)[::-1]
# age_gap=call_record[call_record.age==line].groupby("age_Gap")["time_count"].sum()
# age_gap.median()
def get_Total_Distribution(df):
    age_gap=df.groupby("age_Gap")["time_count"].sum()
    age_gap=age_gap/age_gap.sum()
    ax=age_gap.cumsum().plot(figsize=[12,10],kind="area",rot=50,fontsize=10,colormap='coolwarm')

    
def getGapPlot(df,num=0):
    a=231
    for line in pos[num]:
        plt.subplot(a)
        age_gap=df[df.age==line].groupby("age_Gap_cut")["time_count"].sum()
        age_gap=age_gap/age_gap.sum()
        age_gap=age_gap.cumsum()
        ax=age_gap.plot(figsize=[40,25],kind='bar',rot=60,fontsize=15,colormap='coolwarm')
        ax.text(10,0.5,line,ha="center",color="blue",size=30)
        ax.axhline(0.5)
        a+=1
        
        
get_Total_Distribution(call_record)
```

### 1.从不同年龄差所占的比率分析

#### 1.1  累计分布  瀑布图

```python
def getWF_some_age(year=1990):
    age_gap=call_record[call_record.age==year].groupby("age_Gap")["time_count"].sum()
    age_gap=age_gap/age_gap.sum()
    age_gap=age_gap.sort_values(ascending=False)
    # blank=age_gap.shift(1).fillna(0)
    # step = blank.reset_index(drop=True).repeat(3).shift(-1)
    # step[1::3] = np.nan
    blank=age_gap.cumsum().shift(1).fillna(0)
    ax=age_gap.plot.bar(stacked=True,bottom=blank,cmap='jet',fontsize=12,rot=60,figsize=[25,20])
    ax.text(10,0.5,year,ha="center",color="blue",size=30)
    return age_gap.cumsum()
# ax.plot(step.index, step.values,'k')

plt.subplot(221)
age95=getWF_some_age(1995)
plt.subplot(222)
age63=getWF_some_age(1963)
plt.subplot(223)
age85=getWF_some_age(1985)
plt.subplot(224)
age73=getWF_some_age(1973)
```

通过瀑布图可以看出，不同年龄的人群电话联系人的分布是不一样的，比如1963年的人联系人的年龄差分布比较均匀或者离散程度比较大，或者是不集中，联系人中年龄差的top  11 才达到所有联系人的50%,top3只占有所有联系的18.85%,相比出生于1995年的人的年龄差分布却是差异比较大，或者相对比较集中，联系人中年龄差的top4 已经打到所有联系人的54%，top3占所有联系人的44.89%,其更大的差异在于后者top6的年龄差都在3以内，而前者却相对来说比较离散top6的年龄差分别是：0，-8，-24，-7，-27，-23，所以不同出生年份的人群各个年龄差所在所有联系的人比率以及联系topN年龄差的研究是非常有意义的。

```python
def getStatTop(df,top=6):
    gap_ratio=pd.DataFrame()
    gap=pd.DataFrame()
    for line in pos_span:
        age_gap=df[df.age==line].groupby("age_Gap")["time_count"].sum()
        age_gap=age_gap/age_gap.sum()
        age_gap=age_gap.sort_values(ascending=False)
        age_gap=age_gap.head(top)
        gap_ratio[line]=age_gap.describe()
        gap[line]=age_gap.index.values
    return gap_ratio.T.sort_index(),gap
        
        
gap_r,gap=getStatTop(call_record)
```

由上图我们可以得出以下结论：
-  年龄较小的人联系的人会越紧密，从1965到1995年，top6均值呈现正相关的关系，每个年龄差占所有联系人的平均比率：5%到10%，说明年龄较小的联系的人的紧密程度就会越高，即每个年龄差的人占比平均水平越高
- 年龄越小年龄差的稳定性越小，由上图可知，top6 的方差成正相关关系，随着年龄增大，top6 的方差越来越小，即top6 之间的差异性并不是很大，说明联系人年龄差距的稳定性,下图的极差也同样说明这个问题

```python
def getTop(df,top=3,cumsum=True):
    top3=[]
    for line in pos_span:
   
        age_gap=df[df.age==line].groupby("age_Gap")["time_count"].sum()
        age_gap=age_gap.sort_values(ascending=False)
        if cumsum:
            age_gap=age_gap.cumsum()/age_gap.sum()
        else:
            age_gap=age_gap/age_gap.sum()

        top3.append(age_gap.values[top-1])
      
    return top3

        
plt.plot(pos_span,getTop(call_record,top=6))
```

```python
def  getTop9(df):
    top=pd.DataFrame()
    for i in np.arange(1,10):
        col_name="top"+str(i)
        topn=getTop(df,top=i,cumsum=False)
        top[col_name]=topn

    top.index=pos_span
    return top
```
