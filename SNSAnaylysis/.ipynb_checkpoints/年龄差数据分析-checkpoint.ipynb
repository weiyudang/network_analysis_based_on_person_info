```python
def getGapPlot(df,num=0):
    a=231
    for line in pos[num]:
        plt.subplot(a)
        age_gap=df[df.age==line].groupby("age_Gap")["time_count"].sum()
        age_gap=age_gap/age_gap.sum()
        ax=age_gap.plot.bar(figsize=[40,25],rot=50,fontsize=15)
        ax.text(40,age_gap.quantile(0.98),line,ha="center",color="blue",size=30)
        a+=1

getGapPlot(call_record,num=0)
```

###  对不同年龄差人群，进行统计分析
- 峰度可以度量某一年龄与其联系人年龄差的集中程度，峰值越大说明该年龄接触的联系人越集中
> 峰度（Kurtuosis）是描述总体中所有制分布形态都陡缓成都的统计量。这个统计量需要与正太分布相比较，峰度为0表示该该总体数据分布于正态分布的
陡缓程度相同；**峰度大于0**表示该总体数据分布与正态分布相比较为陡峭，为**尖顶峰**；**峰度小于0**表示该总体数据分布与正态分布相比较为平坦，为**平顶峰**。峰度的绝对值数值越大表示其分布形态的陡缓程度与正态分布的差异程度越大。
- 偏度可以反映不同年龄联系人一种年龄差分布，即：右偏说明该年龄联系人群大部分比他小，左偏说明该
> 偏度与峰度类似，它也是描述数据分布形态的统计量，其描述的是某总体取值分布的对称性。这个统计量同样需要与正态分布相比较，偏度为0表示其数据分布形态与正态分布的偏斜程度相同；偏度大于0表示其数据分布形态与正态分布相比为正偏或右偏，即有一条长尾巴拖在右边，数据右端有较多的极端值；偏度小于0表示其数据分布形态与正态分布相比为负偏或左偏，即有一条长尾拖在左边，数据左端有较多的极端值。偏度的绝对值数值越大表示其分布形态的偏斜程度越大。
- 方差
> 方差反应数据分布的离散程度
- 均值
>  均值反应数据分布的集中程度
- 变异系数


```python
## 获取统计量函数
def get_total_stat(df):
    age_gap=df[df.time_count!=0.0][["time_count","age_Gap"]].values.T
    # age_gap=call_record[call_record.time_count!=0.0][["time_count","age_Gap"]].values.T
    age_gap=np.repeat(age_gap[1],map(int,age_gap[0]))
    kurt=sas.kurtosis(age_gap)
    skew=sas.skew(age_gap)
    age_std=age_gap.std()
    age_mean=age_gap.mean()
    coef=age_gap.std()/age_gap.mean()
    Normtest=sas.kstest(age_gap,"norm").pvalue
    chitest=sas.chisquare(age_gap).pvalue
    stats_df=pd.DataFrame(np.array([kurt,skew,age_mean,age_std,coef,Normtest,chitest]).T,
                          index=["Kurtosis","Skewness","Mean","Standard_Deviation","coef_var","norm_Pvalues","chiPvalue"],columns=["stats"])
    return stats_df.T
def getStats(df):
    kurt=[]
    skew=[]
    age_std=[]
    age_mean=[]
    coef=[]
    Normtest=[]
    chitest=[]
    for born_year in np.arange(1963,1998):
        age_gap=df[(df.age==born_year)&(df.time_count!=0.0)][["time_count","age_Gap"]].values.T
        # age_gap=call_record[call_record.time_count!=0.0][["time_count","age_Gap"]].values.T

        age_gap=np.repeat(age_gap[1],map(int,age_gap[0]))
        kurt.append(sas.kurtosis(age_gap))
        skew.append(sas.skew(age_gap))
        age_std.append(age_gap.std())
        age_mean.append(age_gap.mean())
        coef.append(age_gap.std()/age_gap.mean())
        Normtest.append(sas.kstest(age_gap,"norm").pvalue)
        chitest.append(sas.chisquare(age_gap).pvalue)
    stats_df=pd.DataFrame(np.array([kurt,skew,age_mean,age_std,coef,Normtest,chitest]).T,
                          columns=["Kurtosis","Skewness","Mean","Standard_Deviation","coef_var","norm_Pvalues","chiPvalue"],index=np.arange(1963,1998))
    return stats_df


def getPlot(x,y,ylabel):
    sns.jointplot(x,y,xlim=[1960,2000],size=7,ratio=3,stat_func=sas.spearmanr,\
                  marginal_kws=dict(bins=10, rug=True),s=40,
                  annot_kws=dict(stat="r"),edgecolor="w",space=0.05, linewidth=1).set_axis_labels("Year", ylabel)
sns.set( color_codes=True)

def getRegPlot(x,y,ylabel):
    sns.jointplot(x,y,kind="reg",xlim=[1960,2000],size=7,ratio=3,stat_func=sas.spearmanr,\
                  marginal_kws=dict(bins=10, rug=True),
                  annot_kws=dict(stat="r"),space=0.05).set_axis_labels("Year", ylabel)

    
born_year_range=np.arange(1963,1998)
```
